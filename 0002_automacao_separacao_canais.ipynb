```python
image_files =  np.sort(glob.glob(input_folder + '/*test*.tiff'))
```

```python
image_files
```

```python
for i in image_files:
    image = imread(i)
    ch1 = image[:,:,:,0]
    ch2 = image[:,:,:,1]
    ch3 = image[:,:,:,2]
    
    imsave(os.path.join(output,"ch1_"+ f"{os.path.splitext(os.path.basename(i))[0]}.tiff"), ch1)
    imsave(os.path.join(output,"ch2_"+ f"{os.path.splitext(os.path.basename(i))[0]}.tiff"), ch2)
    imsave(os.path.join(output,"ch3_"+ f"{os.path.splitext(os.path.basename(i))[0]}.tiff"), ch3)
```
